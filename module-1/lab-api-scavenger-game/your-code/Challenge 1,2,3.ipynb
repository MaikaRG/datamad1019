```python
import requests
import os
from dotenv import load_dotenv
load_dotenv()

# https://developer.github.com/v3/
def githubRequestAuthorized(resource):
    authToken = os.getenv("GITHUB_API_TOKEN")
    if not authToken:
        raise ValueError("NECESITAS UN TOKEN")
    else:
        print("We have a github token: ", authToken[0:4])
    headers = {
        "Authorization": "token {}".format(authToken)
    }
    url = "https://api.github.com{}".format(resource)
    print("Requesting authorized {}".format(url))
    res = requests.get(url, headers=headers)
    return res
```

```python
contenido = githubRequestAuthorized("/repos/ironhack-datalabs/scavenger/contents").json()
contenido
```

```python
path=[]
for e in contenido:
    if e['path'] != '.gitignore':
        path.append(e["path"])
path
```

```python
listPathCont=[]
for e in path:
    pathCont = githubRequestAuthorized("/repos/ironhack-datalabs/scavenger/contents/{}".format(e)).json()
    listPathCont.append(pathCont)
```

```python
listaArchivos=[]
listaPath=[]
for e in listPathCont:
    for i in e:
        listaPath.append(i['path'])
listaPath
```

```python
import re
scavengerhunt=[]
for e in listaPath:
    if re.search(".scavengerhunt",e):
        scavengerhunt.append(e.split("/"))
scavengerhunt
```

```python
KeysSorted=sorted(scavengerhunt, key= lambda x: x[1])
KeysSorted
```

```python
UrlFinal=[]
for e in KeysSorted:
    pathContTotal = githubRequestAuthorized("/repos/ironhack-datalabs/scavenger/contents/{}/{}".format(e[0],e[1])).json()
    UrlFinal.append(pathContTotal)
UrlFinal
```

```python
fraseEncode=[]
for e in UrlFinal:
    fraseEncode.append(e['content'])
fraseEncode
```

```python
import base64
```

```python
DesC = []
for e in fraseEncode:
    DesC.append(base64.b64decode(e))
DesC
```

```python
UTF8 = []
for e in DesC:
    UTF8.append(e.decode('UTF-8'))
UTF8
```

```python
FIN=[]
for e in UTF8:
    e = re.sub("\n","", e)
    FIN.append(e)
' '.join(FIN)
```
