## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd
```

```python
data = pd.read_csv('../weather-raw.csv')
data.head(22)
```

```python
print(data.shape)
```

```python
null_cols = data.isnull().sum()
null_cols
```

```python
drop_cols = list(null_cols[null_cols == 22].index)
data = data.drop(drop_cols, axis=1)
```

```python
data.shape
```

```python
data1= data.T
data1
```

```python
melted_cols = data.columns[4:-1]
x=data.melt(id_vars=["year","month",'element'],value_vars=melted_cols)
x.head(26)
```

```python
y=x.groupby(['year','month','element']).mean()
display(y.head())
z=y.groupby(['month']).mean()
display(z)
z.rename(columns={'value':'tmedia'},inplace=True)
display(z)
z.T
```
